```python
import torch
from transformers import GPT2Config, GPT2LMHeadModel, GPT2TokenizerFast

plm = 'skt/kogpt2-base-v2'
```

```python
config = GPT2Config.from_pretrained(plm)
print(config)
```

```python
model = GPT2LMHeadModel.from_pretrained(plm, config=config)
print(model)
```

```python
tokenizer = GPT2TokenizerFast.from_pretrained(plm, bos_token='<s>', eos_token='</s>', unk_token='<unk>')
```

```python
text = '근육이 커지기 위해서는'
```

```python
# tokenize: string -> subword
subwords = tokenizer.tokenize(text)
print(subwords)
# convert_tokens_to_ids: subword -> id
input_ids = tokenizer.convert_tokens_to_ids(subwords)
print(input_ids)
```

```python
# encode: string -> id
input_ids = tokenizer.encode(text)
print(input_ids)
```

```python
gen_ids = model.generate(torch.tensor([input_ids]),
                         max_length=128,
                         repetition_penalty=2.0,
                         eos_token_id=tokenizer.eos_token_id,
                         bos_token_id=tokenizer.bos_token_id)
print(gen_ids)
```

```python
generated = tokenizer.decode(gen_ids[0].tolist())
print(generated)
```

```python
def gpt_inference(prompt='근육이 커지기 위해서는'):
    with torch.no_grad():
        # tokenize: string -> subword -> id
#         subwords = tokenizer.tokenize(prompt)        
#         input_ids = tokenizer.convert_tokens_to_ids(subwords)
        # encode: string -> id
    
        ## 1_tokenize
        input_ids = tokenizer.encode(prompt)
        print('#'*40)
        print('1_tokenize\n%s\n  => %s'%(prompt, input_ids))        
        
        ## 2_Generate gpt-inference
        gen_ids = model.generate(torch.tensor([input_ids]),
                                 max_length=128,
                                 repetition_penalty=2.0,
                                 eos_token_id=tokenizer.eos_token_id,
                                 bos_token_id=tokenizer.bos_token_id)
        print('#'*40)
        print('2_generate\n%s'%(gen_ids))              
        
        ## 3_Detokenize        
        generated = tokenizer.decode(gen_ids[0].tolist())
        print('#'*40)
        print('3_Detokenize\n%s'%(generated))             
        
    return generated
```

```python
generate_text = gpt_inference(prompt='도메인 특화 자연어처리 모델을 개발하기 위해서는')
```

```python
prompt = """
Q:질문내용 여기에 입력하세요
A:"""
generate_text = gpt_inference(prompt=prompt)
```

```python
prompt = """
오늘은 텍스트 생성 모델 GPT에게 향후 자연어처리 공부방법에 대해 물어보겠습니다
Q: 인공지능 자연어처리 수업에 오신걸 환영합니다.
A: """
generate_text = gpt_inference(prompt=prompt)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-small")
model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-small")
```

```python
from transformers import pipeline, Conversation, AutoTokenizer
print('start')
generator = pipeline('conversational', model='microsoft/DialoGPT-small')  # pipeline에서 conversational을 체크하고 모델을 입력하자
for idx in range(3):
    usertext = input(">> User:")
    conversation = Conversation(usertext)

    output = generator(
        conversation, pad_token_id=tokenizer.eos_token_id).generated_responses

    print(">> Bot: %s"%(output))
```
